```python
df = pd.read_csv('pokemon/pokemon_alopez247.csv')
print(df.head(5))
```

```python
from scipy.stats import itemfreq
import numpy as np

# count the number of distinct groups along with the number of Pokemon(s) belonging to that group
# available under Type_1 category. 

type_1 = itemfreq(df.iloc[:,2])
# As itemfreq is depricating, we can use unique one..
# print( np.unique(df.iloc[:,2], return_counts=True) )
type_1_grp = len(type_1)
type_1_names = type_1[:,0]
type_1_count = type_1[:,1]
# print(type_1, type_1_grp, type_1_names, type_1_count)
type_1_grps = np.arange(type_1_grp)
bar_width = 0.5

plt.figure(figsize=(18, 10))
# matplotlib.rcParams['figure.figsize'] = [width, height]
# figure(figsize=(1,1)) would create an inch-by-inch image, which would be 80-by-80 pixels 
# unless you also give a different dpi argument.

plt.bar(type_1_grps, type_1_count, bar_width, alpha=0.5, color='g', label='Pokemon count respective to their Type_1')
plt.legend(loc='best')
plt.xticks(type_1_grps + bar_width/2, type_1_names);
```

```python
# Exercise
# Given score points of Men and Women in a certain game played 5 times with the 
# corresponding errors in the observations:

n_grps = np.arange(5)
bar_width = 0.4
sc_for = ['A', 'B', 'C', 'D', 'E']

men_sc = [20, 30, 10, 50, 90]
err_men_sc = [2, 3, 4, 5, 4]

women_sc = [10, 123, 19, 60, 40]
err_women_sc = [1, 6, 2, 8, 7]
# Plot a bar graph depicting scores for all 5 games along with errors in observation.
bar_width=0.3
plt.figure(figsize=(18, 10))

plt.bar(n_grps-bar_width/2, men_sc, bar_width, label="Men", yerr=err_men_sc, capsize=10)
plt.bar(n_grps+bar_width/2, women_sc, bar_width, label="Women", yerr=err_women_sc, capsize=10)
plt.legend(loc='best')
plt.xticks(n_grps, sc_for);
```

```python
# Advanced bar graph

import random
import numpy as np

clrs = np.linspace(0,1,18)
random.shuffle(clrs)
# print(clrs)

colors = []
for i in range(0, 72, 4):
    idx = np.arange(0, 18, 1)
    random.shuffle(idx)
    r = clrs[idx[0]]
    g = clrs[idx[1]]
    b = clrs[idx[2]]
    a = clrs[idx[3]]
    colors.append([r, g, b, a])
# print(colors)
bar_width = 0.5
plt.figure(figsize=(18,8))
bar_graph = plt.bar(type_1_grps, type_1_count, bar_width, alpha=0.5, color=colors)
plt.legend(bar_graph, type_1_names, bbox_to_anchor=(1, 1))
plt.xlabel('Type')
plt.xticks(type_1_grps, type_1_names)
plt.ylabel('Pokemon Count')
plt.title('Number of Pokemon per Type_1')
plt.grid()
plt.ylim(0,130)
```
